### Atribut data

Beberapa atribut umum yang terdapat dalam dataset glass identification meliputi:

RI (Refractive Index): adalah atribut penting dalam identifikasi kaca yang mengukur seberapa banyak cahaya dibelokkan atau dibengkokkan saat melewati material kaca. Indeks bias merupakan sifat optik fundamental dari material yang menentukan bagaimana cahaya merambat melalui medium tersebut.Indeks bias adalah rasio antara kecepatan cahaya di ruang hampa dengan kecepatan cahaya dalam material n=c/v

Nilai lebih besar dari 1: Menunjukkan bahwa cahaya bergerak lebih lambat dalam material dibandingkan di ruang hampa.
Nilai lebih kecil dari 1: Secara teoretis, hal ini tidak mungkin untuk material transparan normal, tetapi bisa terjadi dalam beberapa kondisi ekstrem seperti metamaterial.

Na (Sodium): dalam konteks komposisi kaca adalah elemen kimia yang memainkan peran penting dalam pembentukan dan sifat-sifat kaca. Sodium biasanya ditemukan dalam bentuk senyawa sodium oksida (NaO) di dalam kaca.

Mg (Magnesium): dalam konteks komposisi kaca adalah elemen kimia yang berperan penting dalam mempengaruhi sifat-sifat fisik dan kimia kaca. Magnesium biasanya ditemukan dalam bentuk senyawa magnesium oksida (MgO) di dalam kaca.
Al (Aluminium): Aluminium (Al) dalam konteks komposisi kaca adalah elemen kimia yang memiliki peran penting dalam meningkatkan sifat fisik dan kimia kaca. Aluminium biasanya ditemukan dalam bentuk senyawa aluminium oksida (AlO) di dalam kaca.

Si (Silicon): Silicon (Si) dalam konteks komposisi kaca adalah elemen kimia yang memainkan peran utama dalam membentuk struktur dasar kaca. Silicon biasanya ditemukan dalam bentuk senyawa silika (SiO) di dalam kaca.

K (Potassium): Potassium (K) dalam konteks komposisi kaca adalah elemen kimia yang berperan penting dalam memodifikasi sifat-sifat fisik dan kimia kaca. Potassium biasanya ditemukan dalam bentuk senyawa potassium oksida (KO) di dalam kaca.

Ca (Calcium): Calcium (Ca) dalam konteks komposisi kaca adalah elemen kimia yang berperan penting dalam meningkatkan stabilitas dan sifat mekanik kaca. Calcium biasanya ditemukan dalam bentuk senyawa kalsium oksida (CaO) di dalam kaca.

Ba (Barium): Barium (Ba) dalam konteks komposisi kaca adalah elemen kimia yang digunakan untuk mengubah sifat optik dan mekanik kaca. Barium biasanya ditemukan dalam bentuk senyawa barium oksida (BaO) di dalam kaca.

Fe (Iron): ron (Fe) dalam konteks komposisi kaca adalah elemen kimia yang mempengaruhi berbagai sifat optik dan mekanik kaca. Iron biasanya ditemukan dalam bentuk senyawa besi oksida (FeO atau FeO) di dalam kaca.

### Tipe data

Objek Data dan Jenis Atribut
Dataset terdiri dari objek data yang mewakili entitas. Dalam basis data penjualan, objek mungkin berupa pelanggan, barang toko, dan penjualan; dalam basis data medis, objek bisa berupa pasien; dalam basis data universitas, objek bisa berupa mahasiswa, profesor, dan mata kuliah. Objek data biasanya dijelaskan oleh atribut dan dapat disebut sebagai sampel, contoh, instance, titik data, atau objek. Jika objek data disimpan dalam basis data, mereka adalah data tuplebaris dalam basis data sesuai dengan objek data, dan kolom sesuai dengan atribut.

Apa Itu Atribut?
Atribut adalah field data yang mewakili karakteristik atau fitur dari objek data. Istilah atribut, dimensi, fitur, dan variabel sering digunakan secara bergantian. Atribut yang menjelaskan objek pelanggan bisa meliputi, misalnya, ID pelanggan, nama, dan alamat. Nilai yang diamati untuk atribut tertentu dikenal sebagai observasi. Satu set atribut yang digunakan untuk menjelaskan objek tertentu disebut sebagai vektor atribut atau vektor fitur. Jenis atribut ditentukan oleh set nilai yang mungkinnominal, biner, ordinal, atau numerik.

Atribut Nominal
Nilai atribut nominal adalah simbol atau nama yang tidak memiliki urutan yang berarti. Contoh atribut nominal termasuk warna rambut (hitam, coklat, pirang) dan status pernikahan (single, menikah, bercerai). Atribut nominal juga dikenal sebagai atribut kategorikal.

Atribut Biner
Atribut biner adalah atribut nominal dengan dua kategori atau keadaan: 0 atau 1. Atribut ini disebut boolean jika dua keadaan tersebut sesuai dengan benar dan salah. Contoh atribut biner adalah perokok (1 jika pasien merokok, 0 jika tidak) dan hasil tes medis (1 untuk positif, 0 untuk negatif). Atribut biner simetris jika kedua keadaan sama pentingnya. Atribut biner asimetris jika hasilnya tidak sama pentingnya.

Atribut Ordinal
Atribut ordinal memiliki nilai yang memiliki urutan yang bermakna, tetapi tidak diketahui besarnya perbedaan antara nilai berturut-turut. Contoh atribut ordinal termasuk ukuran minuman (kecil, sedang, besar) dan peringkat profesional. Atribut ordinal berguna untuk menilai kualitas yang tidak bisa diukur secara objektif, seperti kepuasan pelanggan.

Atribut Numerik
Atribut numerik adalah atribut kuantitatif yang diukur dalam nilai integer atau real. Atribut numerik bisa berupa skala interval atau skala rasio.

Atribut Skala Interval: Diukur pada skala unit berukuran sama dan dapat memiliki nilai positif, 0, atau negatif. Contoh: suhu (derajat Celsius atau Fahrenheit), tanggal kalender. Nilai tidak memiliki titik nol sejati.
Atribut Skala Rasio: Memiliki titik nol sejati dan nilai dapat diukur dalam rasio. Contoh: suhu dalam Kelvin, berat, tinggi, jumlah kata.
Atribut Diskrit vs. Kontinu
Atribut diskrit memiliki set nilai yang terbatas atau dapat dihitung tak terbatas, seperti warna rambut atau ukuran minuman. Atribut kontinu tidak diskrit dan biasanya diwakili sebagai variabel floating-point.

Jenis atribut ini tidak saling eksklusif dan sering diproses berbeda tergantung pada algoritma klasifikasi.

# Modeling

Modeling dalam machine learning adalah proses menciptakan dan melatih algoritma yang dapat mengenali pola dalam data dan membuat prediksi atau keputusan berdasarkan pola tersebut. Proses ini melibatkan penggunaan teknik statistik dan komputasi untuk mengembangkan model yang dapat secara akurat merepresentasikan hubungan dalam data dan memprediksi hasil baru yang belum pernah dilihat sebelumnya. karena didalam kasus Datasets Raisin ini tujuannya adalah untuk menentukan kategori dari pengelompokan jenis kismis jadi tentunya kita menggunakan Metode Klasifikasi

## Pemilihan metode model

Metode klasifikasi adalah teknik analisis data yang digunakan untuk memisahkan atau mengelompokkan data ke dalam kategori atau kelas berdasarkan atribut-atribut yang ada. Tujuan utama dari metode ini adalah membangun model yang dapat memprediksi kelas atau label data yang belum pernah dilihat sebelumnya berdasarkan fitur-fitur yang diamati. Dalam hal ini, kita menggunakan metode klasifikasi K-Nearest Neighbors

# K-Nearest Neighbors

KNN (K-Nearest Neighbors) classifier adalah algoritma dalam machine learning yang digunakan untuk klasifikasi dan regresi. KNN adalah metode non-parametrik dan berbasis instance, yang berarti ia tidak membuat asumsi eksplisit tentang distribusi data dan mendasarkan prediksi pada kasus-kasus terdekat dalam dataset.

Berikut adalah konsep dan langkah-langkah dasar dalam KNN classifier:

Training Phase: Tidak ada fase pelatihan eksplisit dalam KNN. Algoritma hanya menyimpan seluruh dataset pelatihan dan menggunakan dataset ini untuk membuat prediksi.

Prediksi untuk Data Baru:

1.Menghitung Jarak: Untuk setiap data baru yang ingin diprediksi, KNN menghitung jarak antara data baru tersebut dan semua titik data dalam dataset pelatihan. Jarak yang paling umum digunakan adalah jarak Euclidean, tetapi jarak lainnya seperti Manhattan atau Minkowski juga bisa digunakan.
2.Menentukan Tetangga Terdekat: Setelah menghitung jarak, KNN mengidentifikasi k titik data terdekat (neighbor) dari data baru. Parameter
k adalah jumlah tetangga yang akan dipertimbangkan, dan harus ditentukan sebelum menjalankan algoritma.
Voting Mayoritas (Klasifikasi): Dalam tugas klasifikasi, KNN menentukan kelas dari data baru berdasarkan kelas mayoritas dari k tetangga terdekat. Misalnya, jika k=3 dan dua dari tiga tetangga terdekat milik kelas A, maka data baru akan diklasifikasikan sebagai kelas A.
Rata-rata (Regresi): Dalam tugas regresi, KNN menghitung rata-rata dari nilai target dari k tetangga terdekat untuk memprediksi nilai target data baru.

# Stacking Classifier

Stacking classifier, atau stacking ensemble, adalah teknik dalam machine learning yang menggabungkan prediksi dari beberapa model (disebut juga base models atau base learners) untuk meningkatkan kinerja prediksi. Ide utama dari stacking adalah menggunakan model meta (meta-model atau meta-learner) yang belajar bagaimana menggabungkan prediksi dari base models untuk menghasilkan prediksi akhir yang lebih akurat.

Berikut adalah langkah-langkah umum dalam stacking classifier:

1.Membangun Base Models: Beberapa model dasar dilatih menggunakan data pelatihan. Base models ini bisa beragam jenis, seperti decision trees, support vector machines, neural networks, dan sebagainya.

2.Menghasilkan Prediksi Base Models: Setelah base models dilatih, mereka digunakan untuk memprediksi data pelatihan dan data pengujian. Prediksi yang dihasilkan oleh setiap base model pada data pelatihan digunakan sebagai fitur baru untuk pelatihan meta-model.

3.Membangun Meta-Model: Sebuah model meta dilatih menggunakan prediksi dari base models sebagai input (fitur). Meta-model belajar untuk menggabungkan prediksi dari base models untuk membuat prediksi akhir yang lebih baik.

4.Prediksi Akhir: Pada tahap prediksi, base models digunakan untuk menghasilkan prediksi pada data baru, kemudian meta-model menggabungkan prediksi ini untuk menghasilkan prediksi akhir.

```
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, Normalizer

# Membaca data
df = pd.read_excel('glass.xlsx')

# Memisahkan fitur dan label
feature_columns = ['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']
X = df[feature_columns].values
y = df['Type'].values

# Mengencode label
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Normalisasi fitur
scaler = Normalizer()
X = scaler.fit_transform(X)

# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Membuat classifier dengan jumlah tetangga optimal
classifier = KNeighborsClassifier(n_neighbors=2)

# Melatih model dengan data latih
classifier.fit(X_train, y_train)

# Memprediksi data latih
c1tr = classifier.predict(X_train)
accuracy_train = accuracy_score(y_train, c1tr)
print(f'Akurasi data latih: {accuracy_train:.2f}')

# Memprediksi data uji
c1tes = classifier.predict(X_test)
accuracy_test = accuracy_score(y_test, c1tes)
print(f'Akurasi data uji: {accuracy_test:.2f}')

# Memprediksi data baru
x_baru = [[1.51934, 13.64, 3.54, 0.75, 72.65, 0.16, 8.89, 0.15, 0.24]]
x_baru_normalized = scaler.transform(x_baru)
c1baru = classifier.predict(x_baru_normalized)
print("Prediksi untuk data inputan baru:", label_encoder.inverse_transform(c1baru), "atau", c1baru)
```

## Stacking Classifier menggunakan Library skylearn

```
import pandas as pd
import numpy as np
from sklearn.preprocessing import RobustScaler, PowerTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import StackingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder

# Membaca data
df = pd.read_csv('glass.csv')

# Memisahkan fitur dan label
feature_columns = ['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']
X = df[feature_columns].values
y = df['Type'].values

# Mengencode label kelas
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Membagi dataset menjadi set pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Membuat pipeline untuk preprocessing
preprocessor = Pipeline(steps=[
    ('scaler', RobustScaler()),
    ('power_transform', PowerTransformer())
])

# Transformasi data latih dan uji
X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)

# Membuat daftar estimator untuk StackingClassifier
estimators = [(f'KNN {i}', KNeighborsClassifier(n_neighbors=i)) for i in range(1, 30, 1)]

# Inisialisasi StackingClassifier dengan GaussianNB sebagai estimator akhir
clf = StackingClassifier(
    estimators=estimators, final_estimator=GaussianNB()
)

# Mengatur parameter grid untuk tuning
param_grid = {
    'final_estimator__var_smoothing': np.logspace(-9, 0, 10)
}

# Inisialisasi GridSearchCV
grid_search = GridSearchCV(clf, param_grid, cv=5, scoring='accuracy', n_jobs=-1)

# Melatih GridSearchCV
grid_search.fit(X_train_transformed, y_train)

# Menampilkan hasil terbaik
print(f'Best parameters found: {grid_search.best_params_}')
print(f'Best cross-validation accuracy: {grid_search.best_score_:.4f}')

# Mengevaluasi kinerja keseluruhan StackingClassifier setelah tuning
best_clf = grid_search.best_estimator_
overall_accuracy = best_clf.score(X_test_transformed, y_test)
print(f'\nStacking Classifier after tuning:')
print(f'Akurasi: {overall_accuracy:.17f}')

# Membuat prediksi untuk data baru x_baru
x_baru = [[1.51711,14.23,0,2.08,73.36,0,8.62,1.67,0]]
x_baru_transformed = preprocessor.transform(x_baru)  # Transformasi data baru menggunakan preprocessor
prediction = best_clf.predict(x_baru_transformed)
predicted_class = label_encoder.inverse_transform(prediction)
print(f'\nPrediksi x_baru: {predicted_class[0]}')

# Membuat prediksi untuk data baru x_baru menggunakan masing-masing estimator dalam StackingClassifier yang sudah di-fit
individual_estimators = best_clf.named_estimators_
predictions = {name: label_encoder.inverse_transform(estimator.predict(x_baru_transformed))[0] for name, estimator in individual_estimators.items()}
for name, pred in predictions.items():
    print(f"Prediksi dari {name}: {pred}")

# Menghitung jumlah prediksi untuk setiap kelas
unique_classes = label_encoder.classes_
class_counts = {cls: sum(1 for pred in predictions.values() if pred == cls) for cls in unique_classes}

for cls in unique_classes:
    print(f'Jumlah estimator dengan kelas {cls}: {class_counts[cls]}')

# Menentukan kelas berdasarkan jumlah prediksi terbanyak
final_class = max(class_counts, key=class_counts.get)
print(f'\nKelas akhir yang dipilih berdasarkan suara terbanyak: {final_class}')
```

## Random Forest Classifier

Random Forest Classifier adalah salah satu algoritma dalam machine learning yang digunakan untuk tugas klasifikasi dan regresi. Ini termasuk dalam kategori ensemble learning, di mana model dibangun dari beberapa model lain yang disebut sebagai pohon keputusan (decision trees). Berikut adalah penjelasan mengenai Random Forest Classifier:

Ensemble Learning: Random Forest bekerja dengan cara menggabungkan beberapa pohon keputusan yang independen satu sama lain. Setiap pohon keputusan dibuat secara acak dengan menggunakan subset acak dari data training dan subset acak dari fitur (variabel prediktor).

Pohon Keputusan: Pohon keputusan adalah struktur berhirarki yang terdiri dari simpul dan cabang. Setiap simpul dalam pohon mewakili titik pemisahan berdasarkan nilai dari suatu fitur, sedangkan cabang-cabang mewakili hasil dari pemisahan tersebut.

```
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Membaca data
df = pd.read_excel('glass.xlsx')

# Memisahkan fitur dan label
feature_columns = ['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']
X = df[feature_columns].values
y = df['Type'].values

# Mengencode label
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Normalisasi fitur
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Membagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Membuat classifier Random Forest
classifier = RandomForestClassifier(n_estimators=200,random_state=123)

# Melatih model dengan data latih
classifier.fit(X_train, y_train)

# Memprediksi data latih
c1tr = classifier.predict(X_train)
accuracy_train = accuracy_score(y_train, c1tr)
print(f'Akurasi data latih: {accuracy_train:.2f}')

# Memprediksi data uji
c1tes = classifier.predict(X_test)
accuracy_test = accuracy_score(y_test, c1tes)
print(f'Akurasi data uji: {accuracy_test:.2f}')

# Memprediksi data baru
x_baru = [[1.51934,13.64,3.54,0.75,72.65,0.16,8.89,0.15,0.24]]
x_baru_normalized = scaler.transform(x_baru)
c1baru = classifier.predict(x_baru_normalized)
print("Prediksi untuk data inputan baru:", label_encoder.inverse_transform(c1baru), "atau", c1baru)
```

# Bagging Classifier

Bagging (Bootstrap Aggregating) adalah teknik ensemble dalam machine learning yang bertujuan untuk meningkatkan stabilitas dan akurasi model prediksi dengan mengurangi varians. Bagging bekerja dengan membuat beberapa subset dari data pelatihan asli menggunakan bootstrap sampling, melatih model yang sama pada setiap subset, dan kemudian menggabungkan prediksi dari semua model untuk menghasilkan prediksi akhir.

Berikut adalah langkah-langkah umum dalam bagging ensemble classifier:

1.Bootstrap Sampling: Dari dataset pelatihan asli, beberapa subset dibuat dengan cara sampling dengan penggantian (replacement). Artinya, setiap subset bisa mengandung duplikat dari beberapa contoh data dan tidak mengandung beberapa contoh data lain dari dataset asli.

2.Training Base Models: Model yang sama dilatih secara terpisah pada setiap subset data yang dihasilkan dari bootstrap sampling. Karena setiap subset berbeda, masing-masing model akan belajar dari data yang sedikit berbeda.

3.Menggabungkan Prediksi: Setelah semua model dasar dilatih, mereka digunakan untuk memprediksi data baru. Prediksi dari semua model ini kemudian digabungkan untuk menghasilkan prediksi akhir. Untuk tugas regresi, prediksi akhir biasanya merupakan rata-rata dari semua prediksi model. Untuk tugas klasifikasi, prediksi akhir biasanya merupakan hasil voting mayoritas dari semua prediksi model.

```
import matplotlib.pyplot as plt
from sklearn.ensemble import BaggingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import pandas as pd

# Membaca data
df = pd.read_excel('glass.xlsx')

# Memisahkan fitur dan label
feature_columns = ['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']
X = df[feature_columns].values
y = df['Type'].values

# Mengencode label kelas
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

# Normalisasi data fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Membagi data menjadi set pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_encoded, test_size=0.2, random_state=0)

# Inisialisasi list untuk menyimpan hasil
accuracies = []

# Loop untuk nilai n_estimators dari 3 hingga 19
for n in range(3, 20):
    # Inisialisasi BaggingClassifier dengan GaussianNB sebagai estimator dasar
    clf = BaggingClassifier(base_estimator=KNeighborsClassifier(n_neighbors=3), n_estimators=n, random_state=0)

    # Latih model dengan data pelatihan
    clf.fit(X_train, y_train)

    # Prediksi untuk data pengujian
    y_pred = clf.predict(X_test)

    # Hitung akurasi
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append((n, accuracy))
    print(f'n_estimators={n}, Akurasi: {accuracy:.2f}')

# Memisahkan nilai n_estimators dan akurasi untuk plotting
n_estimators, accuracy_values = zip(*accuracies)

# Membuat plot
plt.figure(figsize=(10, 6))
plt.plot(n_estimators, accuracy_values, marker='o', linestyle='-', color='b')
plt.xlabel('Jumlah Estimator')
plt.ylabel('Akurasi')
plt.title('Akurasi vs. Jumlah Estimator pada BaggingClassifier dengan GaussianNB')
plt.grid(True)
plt.show()

# Normalisasi data baru sebelum prediksi
x_baru = [[1.51711,14.23,0,2.08,73.36,0,8.62,1.67,0]]
x_baru_scaled = scaler.transform(x_baru)
prediction = clf.predict(x_baru_scaled)
predicted_class = label_encoder.inverse_transform(prediction)
print(f'Prediksi untuk input baru: {predicted_class[0]}')
```
